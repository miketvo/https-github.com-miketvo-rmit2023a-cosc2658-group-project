# Algorithm Complexity Analysis:

Linear Character Swap - Breadth First vs. Depth First for secret key guessing

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

## Time complexity

This is the time complexity test results for auto (best) algorithm selection. Since we can see that both of them are O(n) based on [their code](../../src/main/java/vn/rmit/cosc2658/development/mike_algo1/SecretKeyGuesser.java), the overall time complexity should also be O(n).

Test code (abbreviated):

```java
    final int MAX_KEY_LENGTH = 512;

    for (int keyLength = 1; keyLength <= MAX_KEY_LENGTH; keyLength++) {
        SecretKey sk = new SecretKey(keyLength, 0);  // Seed = 0 to ensure reproducible results
        SecretKeyGuesser.start(sk, keyLength, SecretKeyGuesser.Algorithm.(...), false);
    }
```

````python
timeData = {
    "Auto": pd.read_csv("../../testData/mikeAlgo1/randomKeyVariableLengthTestAuto.csv"),
    "DepthFirst": pd.read_csv("../../testData/mikeAlgo1/randomKeyVariableLengthTestDepthFirst.csv"),
    "BreadthFirst": pd.read_csv("../../testData/mikeAlgo1/randomKeyVariableLengthTestBreadthFirst.csv")
}
````

````python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(14, 7))

sns.lineplot(x=timeData["Auto"]["KeyLength"], y=timeData["Auto"]["RunTime"], ax=ax[0], c="crimson")
sns.lineplot(x=timeData["Auto"]["KeyLength"], y=timeData["Auto"]["GuessCount"], ax=ax[1], c="dodgerblue")
fig.suptitle("Auto-Best Algorithm\nRuntime and Guess Count vs. Key Length", fontsize=18, fontweight="bold")

plt.show()
fig.savefig("images/TimeComplexity-AlgoAutoBest.png")
````

````python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(14, 7))

sns.lineplot(x=timeData["DepthFirst"]["KeyLength"], y=timeData["DepthFirst"]["RunTime"], ax=ax[0], c="crimson")
sns.lineplot(x=timeData["DepthFirst"]["KeyLength"], y=timeData["DepthFirst"]["GuessCount"], ax=ax[1], c="dodgerblue")
fig.suptitle("Depth First Algorithm\nRuntime and Guess Count vs. Key Length", fontsize=18, fontweight="bold")

plt.show()
fig.savefig("images/TimeComplexity-AlgoDepthFirst.png")
````

````python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(14, 7))

sns.lineplot(x=timeData["BreadthFirst"]["KeyLength"], y=timeData["BreadthFirst"]["RunTime"], ax=ax[0], c="crimson")
sns.lineplot(x=timeData["BreadthFirst"]["KeyLength"], y=timeData["BreadthFirst"]["GuessCount"], ax=ax[1], c="dodgerblue")
fig.suptitle("Breadth First Algorithm\nRuntime and Guess Count vs. Key Length", fontsize=18, fontweight="bold")

plt.show()
fig.savefig("images/TimeComplexity-AlgoBreadthFirst.png")
````
